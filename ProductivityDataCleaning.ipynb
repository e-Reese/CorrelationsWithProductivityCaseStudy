```python
# Importing libraries
import numpy as np
import pandas as pd
```

```python
# Loading in productivity data
gdpPerHoursWorked = pd.read_csv('RawData/gdp_over_hours_worked_with_estimated_hours_worked.csv')
```

```python
# Cut down data to desired years, desired countries, and needed columns. Find definitions for column names
countryCodes = [
    'AUS', 'AUT', 'BEL', 'CAN', 'HRV', 
    'CZE', 'DNK', 'EST', 'FIN', 'FRA', 
    'DEU', 'GRC', 'HUN', 'ISL', 'ISR', 
    'ITA', 'JPN', 'KOR', 'LVA', 'LTU', 
    'LUX', 'NLD', 'NZL', 'NOR', 'POL', 
    'PRT', 'SVK', 'SVN', 'ESP', 'SWE', 
    'CHE', 'GBR', 'USA'
]
gdpPerHourFilter = gdpPerHoursWorked[(gdpPerHoursWorked['year'] > 2006) & (gdpPerHoursWorked['iso3c'].isin(countryCodes))]
```

```python
gdpPerHourLessColumns = gdpPerHourFilter.drop(['pop_0_to_14', 'pop_15_to_64', 'labor_force_partic_ILO', 'labor_force_partic_ILO_female', 
                                               'pop_percent_male', 'hour_worked_adjustment', 'hours_worked_adjustment', 'gdp_ppp_over_pop_adjusted_for_hours', 
                                               'gdp_ppp_over_population_15_to_65', 'iso2c', 'penn_pop', 'penn_employment', 'penn_average_hours_worked', 
                                               'penn_employment_prop', 'total_hours_alternative', 'oil', 'use_model', 'estimated_using_past_value', 
                                               'estimated_using_model', 'is_grouping', 'hours_worked_KNOWN_PLUS_ESTIMATED', 'region', 'pop_over_65',
                                               'working_age_pop_pct', 'employment_rate', 'employed', 'unemployment_r' ], axis=1)
```
