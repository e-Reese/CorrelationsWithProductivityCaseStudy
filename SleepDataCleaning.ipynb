```python
#importing the libraries
import numpy as np
import pandas as pd
```

```python
''' 
Legend:
TST = Total Sleep Time measured in minutes
TIB = Time In Bed (measured in minutes)
SE = Sleep Efficiency (calculated as (TST/TIB)*100)
SOT = Sleep Onset Time 
WT = Wake Time (time of awaking)
SOL = Sleep Onset Latency 
WASO = Wake After Sleep Onset
'''
```

```python
#loading sleep data
weekdaySleep = pd.read_csv('SleepMeasurementData-WeekdaySleep_WithEfficiency.csv')
weekendSleep = pd.read_csv('SleepMeasurementData-WeekendSleep_WithEfficiency.csv')
```

```python
weekdaySleep.head()
```

```python
weekendSleep.head()
```

```python
weekendSleep.drop(['Country'], axis=1, inplace=True)
#Combine both dataframes
combinedSleep = pd.concat([weekdaySleep, weekendSleep], axis=1)
```

```python
#Split SD into separate row
combinedSleep['TSTwd (SD)'] = combinedSleep['TSTwd (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
combinedSleep[['TSTwd', 'TSTwd SD']] = combinedSleep['TSTwd (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
combinedSleep.drop(['TSTwd (SD)'], axis=1, inplace=True)
```

```python
#Split SD into separate row
combinedSleep['SOLwd (SD)'] = combinedSleep['SOLwd (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
combinedSleep[['SOLwd', 'SOLwd SD']] = combinedSleep['SOLwd (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
combinedSleep.drop(['SOLwd (SD)'], axis=1, inplace=True)
```

```python
combinedSleep['SEwd/% (SD)'] = combinedSleep['SEwd/% (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
combinedSleep[['SEwd/%', 'SEwd/% SD']] = combinedSleep['SEwd/% (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
combinedSleep.drop(['SEwd/% (SD)'], axis=1, inplace=True)

combinedSleep['WASOwd (SD)'] = combinedSleep['WASOwd (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
combinedSleep[['WASOwd', 'WASOwd SD']] = combinedSleep['WASOwd (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
combinedSleep.drop(['WASOwd (SD)'], axis=1, inplace=True)

combinedSleep['TSTwe (SD)'] = combinedSleep['TSTwe (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
combinedSleep[['TSTwe', 'TSTwe SD']] = combinedSleep['TSTwe (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
combinedSleep.drop(['TSTwe (SD)'], axis=1, inplace=True)

combinedSleep['SOLwe (SD)'] = combinedSleep['SOLwe (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
combinedSleep[['SOLwe', 'SOLwe SD']] = combinedSleep['SOLwe (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
combinedSleep.drop(['SOLwe (SD)'], axis=1, inplace=True)

combinedSleep['SEwe/% (SD)'] = combinedSleep['SEwe/% (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
combinedSleep[['SEwe/%', 'SEwe/% SD']] = combinedSleep['SEwe/% (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
combinedSleep.drop(['SEwe/% (SD)'], axis=1, inplace=True)

combinedSleep['WASOwe (SD)'] = combinedSleep['WASOwe (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
combinedSleep[['WASOwe', 'WASOwe SD']] = combinedSleep['WASOwe (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
combinedSleep.drop(['WASOwe (SD)'], axis=1, inplace=True)
```

```python
#Average corresponding weekend and weekday columns
```

```python
combinedSleep.head()
```

```python
combinedSleep['TSTavg'] = combinedSleep[['TSTwd', 'TSTwe']].astype(float).mean(axis=1)
combinedSleep['TST SD avg'] = combinedSleep[['TSTwd SD', 'TSTwe SD']].astype(float).mean(axis=1)
combinedSleep['SOLavg'] = combinedSleep[['SOLwd', 'SOLwe']].astype(float).mean(axis=1)
combinedSleep['SOL SD avg'] = combinedSleep[['SOLwd SD', 'SOLwe SD']].astype(float).mean(axis=1)
combinedSleep['SEavg'] = combinedSleep[['SEwd/%', 'SEwe/%']].astype(float).mean(axis=1)
combinedSleep['SE SD avg'] = combinedSleep[['SEwd/% SD', 'SEwe/% SD']].astype(float).mean(axis=1)
combinedSleep['WASOavg'] = combinedSleep[['WASOwd', 'WASOwe']].astype(float).mean(axis=1)
combinedSleep['WASO SD avg'] = combinedSleep[['WASOwd SD', 'WASOwe SD']].astype(float).mean(axis=1)
```

```python
combinedSleep.drop(['TSTwe', 'TSTwe SD', 'SOLwe', 'SOLwe SD', 'SEwe/%', 'SEwe/% SD', 'WASOwe', 'WASOwe SD'], axis=1, inplace=True)
```

```python
combinedSleep.drop(['TSTwd', 'TSTwd SD', 'SOLwd', 'SOLwd SD', 'SEwd/%', 'SEwd/% SD', 'WASOwd', 'WASOwd SD'], axis=1, inplace=True)
```

```python
combinedSleep.rename(columns={'TST SD avg': 'TSTavgSD', 'SOL SD avg': 'SOLavgSD', 'SE SD avg':' SEavgSD', 'WASO SD avg': 'WASOavgSD'})
```

```python
weekdaySotWt = pd.read_csv('SleepMeasurementData-WeekdaySleep.csv')
weekendSotWt = pd.read_csv('SleepMeasurementData-WeekendSleep.csv')
```

```python
# Splitting and cleaning SOT and WT dataframes
weekdaySotWt['SOTwd (SD)'] = weekdaySotWt['SOTwd (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
weekdaySotWt[['SOTwd', 'SOTwd SD']] = weekdaySotWt['SOTwd (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
weekdaySotWt.drop(['SOTwd (SD)'], axis=1, inplace=True)

weekdaySotWt['WTwd (SD)'] = weekdaySotWt['WTwd (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
weekdaySotWt[['WTwd', 'WTwd SD']] = weekdaySotWt['WTwd (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
weekdaySotWt.drop(['WTwd (SD)'], axis=1, inplace=True)

weekdaySotWt['MSTwd (SD)'] = weekdaySotWt['MSTwd (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
weekdaySotWt[['MSTwd', 'MSTwd SD']] = weekdaySotWt['MSTwd (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
weekdaySotWt.drop(['MSTwd (SD)'], axis=1, inplace=True)


weekendSotWt['SOTwe (SD)'] = weekendSotWt['SOTwe (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
weekendSotWt[['SOTwe', 'SOTwe SD']] = weekendSotWt['SOTwe (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
weekendSotWt.drop(['SOTwe (SD)'], axis=1, inplace=True)

weekendSotWt['WTwe (SD)'] = weekendSotWt['WTwe (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
weekendSotWt[['WTwe', 'WTwe SD']] = weekendSotWt['WTwe (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
weekendSotWt.drop(['WTwe (SD)'], axis=1, inplace=True)

weekendSotWt['MSTwe (SD)'] = weekendSotWt['MSTwe (SD)'].apply(lambda x: str(x).replace('(', '').replace(')', ''))
weekendSotWt[['MSTwe', 'MSTwe SD']] = weekendSotWt['MSTwe (SD)'].apply(lambda x: pd.Series(str(x).split(' ')))
weekendSotWt.drop(['MSTwe (SD)'], axis=1, inplace=True)
```
